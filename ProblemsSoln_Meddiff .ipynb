```python
#problem statement 1 - group files by owners name

def groupFiles(files):
    res = {}
    for file, owner in files.items():  
        res[owner] = res.get(owner, []) + [file]  
    
    return res

files = {
    'Input.txt': 'Randy',
    'Code.py': 'Stan',
    'Output.txt': 'Randy'
}

print(groupFiles(files))
```

```python
#problem statement 3 - parse a log file & extract details of Errors & Warnings recorded into a separate file.

import re


logFile = r"/home/avanseus/mongoLOGS/mongod.log"
exportFile = r"/home/avanseus/mongoLOGS/demologParsed.log"


regex = r'[*WARNING*,*ERROR*]'
read_line = False

with open(logFile, "r") as file:
    match_list = []
    
    data = file.readlines()
    for txt in data:
        if("WARNING" in txt):
            match_list.append("WARNING  -  " +txt.split("WARNING")[-1])
        elif ("ERROR" in txt):
            match_list.append("ERROR  -  " +txt.split("ERROR")[-1])
        
file.close()

with open(exportFile, "w+") as file:
    for txt in match_list:
        
        file.write(txt + "\n")
file.close()
```

```python
#problem statement 4 - display current path
class Path:
    def __init__(self, path):
        self.current_path = path

    def cd(self, new_path):
        i=0;
        new_pathList=new_path.split('/')
        pathLength=len(new_pathList)
        pathList=self.current_path.split('/')
        if new_pathList[0]=='':
            #direct pathname
            del pathList[:]
            pathList.append('/'+new_pathList[1])
            i=i+2
        while(i<pathLength):
            j=len(pathList)-1
            if new_pathList[i]=='..':
                #parent directory
                pathList.pop(j)
            else:
                pathList.append(new_pathList[i])
            i=i+1
        self.current_path="/".join(pathList)
               
        
path = Path('/a/b/c/d')
path.cd('../../x')
print(path.current_path)
```
